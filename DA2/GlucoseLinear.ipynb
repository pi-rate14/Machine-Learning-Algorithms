```python
import pandas as pd #dataset manipulation
import numpy as np #algebraic and arithmetic calculation
import matplotlib.pyplot as plt #data visualization
```

```python
#importing dataset
dataset = pd.read_csv("Glucose_Data.csv")
#iloc- extract the data based on index location
#.values convert into array otherwise it will be as dataframe
x = dataset.iloc[:,0:-1].values
y = dataset.iloc[:,1].values
```

```python
#split dataset as train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.30, random_state=0)
```

```python
#building model
from sklearn.linear_model import LinearRegression
regress = LinearRegression()
regress.fit(x_train, y_train)
```

```python
#predict
y_pred = regress.predict(x_test)
```

```python
plt.title("Simple Linear Regression")
plt.xlabel("Age")
plt.ylabel("Glucose")
#for scattered output
plt.scatter(x_train, y_train, color="red")
#to plot straight line
plt.plot(x_train, regress.predict(x_train), color="blue")
```

```python
#Visualize the output with test data and train data
plt.title("Linear regression with train and test data")
plt.xlabel("Age")
plt.ylabel("Glucose")
plt.scatter(x_test,y_test,color="red")
plt.plot(x_train,regress.predict(x_train),color="blue")
plt.show()
```

```python
def prediction(x):
    return regress.coef_ * x + regress.intercept_
```

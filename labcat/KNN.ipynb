## Q3 (KNN classification on Breast Cancer data set, with accuracy, precision, and recall)

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt #matplotlib is used for plot the graphs,
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv("data.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data=data.drop(["Unnamed: 32","id"],axis=1)
```

```python
data["diagnosis"]=data["diagnosis"].map({'B':0,'M':1}).astype(int)
data.head()
```

```python
corr=data.corr()
corr.nlargest(30,'diagnosis')['diagnosis']
```

```python
x=data[['radius_mean','perimeter_mean','area_mean','compactness_mean','concave points_mean','radius_se','perimeter_se', 'area_se','compactness_se', 'concave points_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','texture_worst','area_worst']]
y=data[['diagnosis']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.72, random_state=1)
```

```python
from sklearn.metrics import accuracy_score
model = KNeighborsClassifier(n_neighbors=8)
model.fit(x_train, y_train)
predict = model.predict(x_test)
accuracy_score(predict,y_test)
```

```python
accuracy=model.score(x_train,y_train)
print(accuracy)
```

```python
model.score(x_train,y_train)
```

```python

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
print("Accuracy Score: ",(accuracy_score(predict,y_test)*100)) 
print ("Precision Score: ",   precision_score(predict,y_test, average='binary')) 
print ("Recall Score: ", recall_score(predict,y_test, average='binary')) 
```
